```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = sns.load_dataset('titanic')
dataset.head()
```

```python
import seaborn as sns
sns.displot(x = dataset['age'], bins = 10)
```

```python
sns.displot(dataset['age'], bins = 10,kde=False)
```

```python
import seaborn as sns
# For Plot 1
sns.jointplot(x = dataset['age'], y = dataset['fare'], kind =
'scatter')
# For Plot 2
sns.jointplot(x = dataset['age'], y = dataset['fare'], kind = 'hex')
```

```python
sns.rugplot(dataset['fare'])
```

```python
sns.barplot(x='sex', y='age', data=dataset)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.barplot(x='sex', y='age', data=dataset, estimator=np.std)
```

```python
sns.countplot(x='sex', data=dataset)
```

```python
sns.boxplot(x='sex', y='age', data=dataset)
```

```python
sns.boxplot(x='sex', y='age', data=dataset, hue="survived")
```

```python
sns.violinplot(x='sex', y='age', data=dataset)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataset = sns.load_dataset('titanic')
dataset.head()
```

```python
numeric_data=dataset.select_dtypes(include=[float,int])
corr = numeric_data.corr()
import seaborn as sns
sns.heatmap(corr)
```

```python
corr = numeric_data.corr()
sns.heatmap(corr, annot=True)
```

```python
corr = numeric_data.corr()
sns.heatmap(corr)
```
